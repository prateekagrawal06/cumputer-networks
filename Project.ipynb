```python
#main function
import copy

def graph_input():
    
    g = []
    file_name = raw_input("\nInput original network topology matix data file: ")
    with open(file_name, "r") as file_rows:
        g = [map(int, line.split()) for line in file_rows]
    print "\nYou entered the following topology"
    s = [[str(e) for e in row] for row in g]
    lens = [max(map(len, col)) for col in zip(*s)]
    fmt = '  '.join('{{:{}}}'.format(x) for x in lens)
    table = [fmt.format(*row) for row in s]
    print '\n'.join(table)
        
        
    return g
    #nodes = len(g[0])
    #source = int(raw_input("Enter the source node"))
    #dijkstra_algo(g, source, nodes)
    
def dijkstra_algo(g, source):
    nodes = len(g[0])
    pre = {}
    dist = {}
    q = []
    for i in range(nodes):
        dist[i] = 999
        pre[i] = [-1]
        q.append(i)
    dist[source] = 0

    while len(q) != 0:
        n_dist = {}
        for i in q:
            n_dist[i] = dist[i]
        u = min(n_dist.items(), key=lambda x: x[1])
        u = u[0]
        q.remove(u)
        for i in range(nodes):
            if g[u][i] not in [0,-1]:
                alt = dist[u] + g[u][i]
                if alt < dist[i]:
                    dist[i] = alt
                    pre[i] = [u]
                elif alt == dist[i]:
                    pre[i].append(u)
    
    print '\nPredecessor Dictionary:'
    print pre
    return pre
                
  
    
    
def print_shortest_path(pre, source,g):
    
    
    nodes = len(g[0])
    a = copy.copy(pre)
    for i in range(nodes):
        l = []

        if i == source:
            continue
        print '\nSmallest path from router', source ,"to router",i
        j = i
        l = [i]
        
        l.append
        
        while j != source:
            l.append(a[j][0])
            j = a[j][0]
        print l[::-1]
        l1 = l[::-1]
        
        cost = calculating_total_cost(l1,g)
        print "Cost = ", cost


def calculating_total_cost(l,g):
    sum1 = 0
    i = 0
    j = 1
    for r in range(len(l)-1):
        sum1 = sum1 + g[l[i]][l[j]]
        i += 1 
        j += 1
    return sum1        

def connection_table(pre, source, g):
    
    
    nodes = len(g[0])
    a = copy.copy(pre)
    c = {}
    c[source] = '-'
    for i in range(nodes):
    
        l = []

        if i == source:
            continue
        j = i
        l = [i]
        while j != source:
            l.append(a[j][0])
            j =a[j][0]
        l = l[::-1]
        
        c[i] = l[1]
    print "\nRouter", source,"Connection Table"
    #print c
    print "{:<8} {:<15}".format('Destination ',' Interface')
    print "======================="
    for k, v in c.iteritems():
        print "{:<8} \t {:<15}".format(k, v)
    
    
    
def changing_weight(g, source):
    edge = map(int,(raw_input("Enter the edge to change the weight(For ex:4 5)").split()))
    weight = int(raw_input("Enter the new weight "))
    g[edge[0]][edge[1]] = weight
    
    print "\nThe modified Network Topology is"
    s = [[str(e) for e in row] for row in g]
    lens = [max(map(len, col)) for col in zip(*s)]
    fmt = '  '.join('{{:{}}}'.format(x) for x in lens)
    table = [fmt.format(*row) for row in s]
    print '\n'.join(table)
    
    pre = dijkstra_algo(g,source)
    
    print_shortest_path(pre, source,g)
    
    connection_table(pre, source,g)

    
def add_node(g, source):

    
    new = map(int,(raw_input("Enter the edge weights from new router to other routers(For ex:4 5 -1...) ").split()))
    
    new.append(0)
    
    for i in range(len(g[0])):
        g[i].append(new[i])
        
    g.append(new)
    
    print "\nThe modified Network Topology is"
    s = [[str(e) for e in row] for row in g]
    lens = [max(map(len, col)) for col in zip(*s)]
    fmt = '  '.join('{{:{}}}'.format(x) for x in lens)
    table = [fmt.format(*row) for row in s]
    print '\n'.join(table)
    
    pre = dijkstra_algo(g,source)
    
    print_shortest_path(pre, source,g)
    
    connection_table(pre, source,g)
    
    
        
    

print "Enter the network"
g = graph_input()



print "\nChoose one if the following options:"
print "1. Find shortest path of source router to all the routers"
print "2. Print connection table for the source router"
print "3. Add an extra edge or change edge weight"
print "4. Add an extra router\n"

inp = int(raw_input())
source = int(raw_input("\nEnter the source router "))

if inp == 1:
   
    
    pre = dijkstra_algo(g,source)
    
    print_shortest_path(pre, source,g)
    
elif inp == 2:
    
    
    
    pre = dijkstra_algo(g,source)
    
    connection_table(pre, source,g)

elif inp == 3:
        
        changing_weight(g, source)
        
elif inp == 4:
    
    
        add_node(g, source)
    
```
